## Import Libraries

```python
import numpy
import math
from matplotlib import pyplot
```

## Define Grid

```python
N=100

x_start, x_end = -5.0, 5.0
y_start, y_end = -2.0, 2.0

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x,y)
```

## Location of vortices

```python
# This shows the x-location of vortices
n_v = 20

xv_list = numpy.linspace(-10,10, n_v)

yv_list=numpy.zeros_like(xv_list)

```

## Strength of Vortices

```python
G = 5.0

gamma = numpy.repeat(G, numpy.size(xv_list))

```

```python
index = numpy.arange(0,len(xv_list))

```

## Define functions to calculate velocities and stream function

```python
def get_velocity_vortices(strength, xv, yv, X, Y):

    u_vortex = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v_vortex = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u_vortex, v_vortex

#def get_streamfunction(strength, xv, yv, X, Y):

   # psi = psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
#    return psi
```

## Calculate velocities and streamfunction

```python

u_vortices = numpy.zeros((N,N))
v_vortices = numpy.zeros((N,N))

#psi_vortices = numpy.zeros((N,N))

for i in index:
    
    u_v, v_v = get_velocity_vortices(gamma[i], xv_list[i], yv_list[i], X, Y)
    
    u_vortices += u_v
    v_vortices += v_v
    
   # psi = get_streamfunction(gamma[i],xv_list[i], yv_list[i], X, Y)
    
   # psi_vortices +=psi
```

## Plot

```python
# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_vortices, v_vortices, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(xv_list, yv_list, color='#CD2305', s=40, marker='o');
```

# Infinite Row of Vortices

Note: 

I thought for a while about whether the locations of the vortices should be included in the equations for velocities.

I'm still not 100 percent convinced whether the velocity equations already account for the locations of the vortices (since a is included in the equations). What I decided to do here was to calculate the velocity field by using two methods and comparing the solutions: 

Method 1 : The velocity field is calculated without accounting for the x, y coordinates of the vortices, and the equations contain only the gridpoints X, Y

Method 2: The velocity field is calculated by including the x, y locations of the vortices, and the equations include 'X-xv' and 'Y-yv' terms.

What I notice is that the velocity fields appear to be very similar except that for method 1 the vortices created by the velocity appear offset from the true locations of the vortices (shown by the red dots in the graphs below). For method 2 I see that the vortices created by the velocity field occur exactly where I have defined them to be. I am not sure if my code has mistakes, but so far this is what I have noticed, and I am inclined to think that method 2 is correct, based on my results.

### Method 1: Calculate the velocity field due to the infinite row of vortices by using mesh points X, Y 

```python
a = (numpy.max(xv_list)-numpy.min(xv_list))/(n_v-1)

gamma2 = numpy.repeat(G, len(X))

u_vortex = + gamma2/(2*a)*(numpy.sinh((2*(Y)*math.pi)/a))/(numpy.cosh((2*(Y)*math.pi)/a)-numpy.cos((2*(X)*math.pi)/a))
v_vortex = - gamma2/(2*a)*(numpy.sin((2*(X)*math.pi)/a))/(numpy.cosh((2*(Y)*math.pi)/a)-numpy.cos((2*(X)*math.pi)/a))
```

```python
# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_vortex, v_vortex, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(xv_list, yv_list, color='#CD2305', s=40, marker='o');
```
