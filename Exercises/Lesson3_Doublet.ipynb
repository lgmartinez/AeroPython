# Doublet

```python
import math
import numpy
from matplotlib import pyplot
```

### Develop Grid

```python
N = 50

x_start, x_end = -2.0, 2.0
y_start, y_end = -2.0, 2.0

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X,Y = numpy.meshgrid(x, y)

print x
```

### Strength and Location of Doublet

```python
strength_doublet = 1.0
x_doublet, y_doublet = 0.0, 0.0
```

#### Define Functions to calculate stream function and velocity components

```python
def get_velocity_doublet(strength, xd, yd, X, Y):
    "Returns the velocity field generated by doublet"
    
    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v

def get_streamfunction_doublet(strength, xd, yd, X, Y):
    "Returns the stream function generated by doublet"
    
    psi = - strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
    
    return psi
```

#### Calculate velocities and stream function using functions defined above:

```python
u_doublet, v_doublet = get_velocity_doublet(strength_doublet, x_doublet, y_doublet, X, Y)

psi_doublet = get_streamfunction_doublet(strength_doublet, x_doublet, y_doublet, X, Y)
```

### Graph!

```python
# Streamline Plot

%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))

pyplot.xlabel('x', fontsize = 16)
pyplot.ylabel('y', fontsize = 16)

pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)

pyplot.streamplot(X, Y, u_doublet, v_doublet, density = 2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='m', s=400, marker='*')

```

# Uniform Flow Past Doublet

#### Define Freestream velocities and streamlines

```python
u_inf = 1.0  # free stream speed

u_freestream = u_inf*numpy.ones((N,N), dtype=float)
v_freestream = numpy.zeros((N,N),dtype=float)

psi_freestream = u_inf * Y
```

#### Superposed velocities and streamlines

```python
u = u_freestream + u_doublet
v = v_freestream + v_doublet

psi = psi_freestream + psi_doublet
```

#### Find stagnation points

```python
# To Find stagnation points, use velocity component u and set y = 0, solve for x:

x_stagn1, y_stagn1 = +math.sqrt(strength_doublet/(2*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(strength_doublet/(2*math.pi*u_inf)), 0
```

## Graph!

```python
# plot the stream lines:

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='r', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='m', s=80, marker='*')

# Add stagnation points:

pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

## Bernoulli's Equation and Pressure Coefficient

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

# plots the pressure coefficient field
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-2.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')
pyplot.contour(X,Y,psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid')
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

### Challenge Task!

PROOF!

We want to start with the equation given for Cp:

Cp = 1 - (V/V_∞)^2

We want to obtain an expression for V and substitute.

We start with the superposed stream line for freestream and doublet:

φ=V_∞*rsin(θ)- κsin(θ)/r

Then at φ=0, we set a=r. We then can solve for the doublet strength:

κ = V_∞ * a^2

Then we plug this into the original equation to obtain an alternative expression for the streamfunction:

φ=V_∞*sin(θ)*[r - (a/r)^2]

Using the definitions for the velocity components:

Uradial = (1/r)* d/dθ(φ) => Ur=V_∞*cos(θ)*[1 - (a/r)^2]

Utangential = -d/dr(φ) => Utangential = -V_∞*sin(θ)*[1 + (a/r)^2]

On the surface of the cylinder, a = r = (strength_doublet)/(2*pi*u_inf), then 

Uradial = 0 and Utangential = -2V_∞*sin(θ)

Finally substituting into Cp:

Cp = 1 - 4*{sin(θ)}^2
