# Source & Sink

## Import Python Libraries

```python
from math import *
import numpy as np
import matplotlib.pyplot as plt
```

## Define Gridpoints

```python
N = 50                        # Total number of gridpoints in both x and y directions
x_start, x_end = -2.0, 2.0   # x goes from -2 to 2
y_start, y_end = -1.0, 1.0   # y goes from -2 to 2

x = np.linspace(x_start, x_end, N)  # This command creates N equidistant gridpoints from x = -2 to x = 2
y = np.linspace(y_start, y_end, N)  # This command creates N equidistant gridpoints from y = -2 to y = 2

X, Y = np.meshgrid(x, y)  # This generates a mesh grid, note that X, Y are different than lowercase x, y!
                          # More on np.meshgrid: http://docs.scipy.org/doc/numpy/reference/generated/numpy.meshgrid.html  
```

## Plot Gridpoints

```python
%matplotlib inline

# The following code plots the grid of points

size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size)) # figsize (w,h, tuple in inches)
plt.xlabel('x', fontsize=16)  # labels the x axis as "x" with font size of 16
plt.ylabel('y', fontsize=16)  # labels the y axis with "y" with font size of 16
plt.xlim(x_start, x_end)      # sets the x-axis from -2 to 2
plt.ylim(y_start, y_end)      # sets the y-axis from -2 to 2
plt.scatter(X, Y, s=10, color='k', marker='o', linewidth=0);  # s = size of dots, k = black color, linewidth

# more on colors: http://matplotlib.org/api/colors_api.html?highlight=color#module-matplotlib.colors
```

## Source Flow

### Define Source Strength and origin, and calculate velocities

```python
strength_source = 5.0            # Defines source strength
x_source, y_source = -1.0, 0.0   # Defines source location

r = np.sqrt(((X-x_source)**2)+((Y-y_source)**2))


# This code computes the x and y components of the radial velocity

u_source = strength_source/(2*pi)*((X-x_source)/(r**2))
v_source = strength_source/(2*pi)*((Y-y_source)/(r**2))
```

### Plot streamlines

```python
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_source, v_source, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_source, y_source, color='m', s=80, marker='o', linewidth=0);
```

# Sink Flow

### Define Sink Strength and origin, and calculate velocities

```python
strength_sink = -5.0
x_sink, y_sink = 1.0, 0.0

# This code computes the x and y components of velocity

u_sink = strength_sink/(2*pi) * (X-x_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
v_sink = strength_sink/(2*pi) * (Y-y_sink)/((X-x_sink)**2 + (Y-y_sink)**2)
```

### Plot streamlines

```python
# This code plots the streamlines of the sink

size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_sink, v_sink, density=2, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter(x_sink, y_sink, color='m', s=80, marker='o', linewidth=0);
```

# Source-Sink Pair

### Velocities of Source-Sink Pair

```python
# This code computes the velocity of the pair source/sink by superposition
u_pair = u_source + u_sink
v_pair = v_source + v_sink
```

### Plot Streamlines of Pair

```python
# This code plots the streamlines of the pair source/sink
size = 10
plt.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.streamplot(X, Y, u_pair, v_pair, density=2.0, linewidth=1, arrowsize=2, arrowstyle='->')
plt.scatter([x_source, x_sink], [y_source, y_sink], color='m', s=80, marker='o', linewidth=0);
```

Moving the source or sink locations and or changing the strengths will alter the streamline plot shown. If say, for example, the source sink is increased and the source remains constant, the streamlines will have a "shorter path" towards the sink (the sinw will have more influence on the flow field). Alternatively if the source is increased and sink remains constant, the streamlines will have a "longer path" towards the sink (the sink will have less influence).

## Challenge Task: Write code to show potential lines

### Velocity Potential Functions

```python
r_sink = np.sqrt(((X-x_sink)**2)+((Y-y_sink)**2))

potential_source = (strength_source/(2*pi)) *  np.log(r)
potential_sink = (strength_sink/(2*pi)) * np.log(r_sink)
potential_pair = potential_source + potential_sink
```

### Plot Potential Functions

```python
# This plots for source and sink of equal strength
# Note that if we change either the source or sink strength, the pair relationship will be different than the "balanced" form


plt.figure(figsize=(20,20))

plt.subplot(3,3,1)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_source)
plt.scatter(x_source, y_source, color='m', s=80, marker='o', linewidth=0);
CS = plt.contour(X, Y, potential_source)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Source');

plt.subplot(3,3,2)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_sink)
plt.scatter(x_sink, y_sink, color='m', s=80, marker='o', linewidth=0);
CS = plt.contour(X, Y, potential_sink)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Sink');

plt.subplot(3,3,3)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_pair)
plt.scatter([x_source, x_sink], [y_source, y_sink], color='m', s=80, marker='o', linewidth=0)
CS = plt.contour(X, Y, potential_pair)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Pair');
```

Similar to the analysis for the streamlines, the potential functions will change when the source or sink strengths are modified. An example, where the sink is increased, is shown below.

```python
strength_sink2 = -10.0

potential_source = (strength_source/(2*pi)) *  np.log(r)
potential_sink = (strength_sink2/(2*pi)) * np.log(r_sink)
potential_pair = potential_source + potential_sink
```

```python
plt.figure(figsize=(20,20))

plt.subplot(3,3,1)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_source)
plt.scatter(x_source, y_source, color='m', s=80, marker='o', linewidth=0);
CS = plt.contour(X, Y, potential_source)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Source');

plt.subplot(3,3,2)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_sink)
plt.scatter(x_sink, y_sink, color='m', s=80, marker='o', linewidth=0);
CS = plt.contour(X, Y, potential_sink)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Sink');

plt.subplot(3,3,3)
plt.xlabel('x', fontsize=16)
plt.ylabel('y', fontsize=16)
plt.xlim(x_start, x_end)
plt.ylim(y_start, y_end)
plt.contour(X, Y, potential_pair)
plt.scatter([x_source, x_sink], [y_source, y_sink], color='m', s=80, marker='o', linewidth=0)
CS = plt.contour(X, Y, potential_pair)
plt.clabel(CS, inline1=1, fontsize=10)
plt.title('Velocity Potential for Pair');
```

This shows that, since the sink is stronger, the potential lines for the source are different than the previous graph. The potential funtion from the sink has more influence on the flow field
