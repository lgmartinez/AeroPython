# Source Distribution on an Airfol

```python
import math
import numpy
from matplotlib import pyplot
```

### Create Grid

```python
N = 51

x_start, x_end = -1.0, 2.0
y_start, y_end = -0.5, 0.5

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X,Y = numpy.meshgrid(x, y)
```

```python
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size)) # figsize (w,h, tuple in inches)
pyplot.xlabel('x', fontsize=16)  # labels the x axis as "x" with font size of 16
pyplot.ylabel('y', fontsize=16)  # labels the y axis with "y" with font size of 16
pyplot.xlim(x_start, x_end)      # sets the x-axis from -2 to 2
pyplot.ylim(y_start, y_end)      # sets the y-axis from -2 to 2
pyplot.scatter(X, Y, s=10, color='k', marker='o', linewidth=0);  # s = size of dots, k = black color, linewidth
```

### Load data on Location and Strength of Sources

```python
xs_airfoil = numpy.loadtxt('NACA0012_x.txt', dtype = float)
ys_airfoil = numpy.loadtxt('NACA0012_y.txt', dtype = float)
strength_sources = numpy.loadtxt('NACA0012_sigma.txt', dtype = float)
```

### Define Freestream velocity and Streamfunction

```python
u_inf = 1.0 # freestream velocity
psi_freestream = u_inf*y #freestream streamline
```

### Define the index for array of number of sources

```python
# Note that I have isued "len(xs_airfoil)", although I could have also used "len(ys_airfoil)" since what we are really after 
# is defining the length of the source array and xs_airfoil = ys_airfoil 

# This index is used to calculate the velocities and streamfunction below
index = range(0, len(xs_airfoil))
```

### Define Functions to calculate stream function and velocity components

```python
#Define Functions to calculate stream function and velocity components

def get_velocity_sources(strength, xs, ys, X, Y):
    "Returns the velocity field generated by sources"
     
    u_source = strength/(2*math.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v_source = strength/(2*math.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
      
    return u_source, v_source

def get_streamfunction_sources(strength, xs, ys, X, Y):
    "Returns the stream function generated by sources"
    
    psi = (strength/(2*math.pi))*numpy.arctan2((y-ys), (x-xs))
    
    return psi
```

### Calculate velocity components and streamfunction

```python
# calculate velocities:

u_sources = numpy.zeros((N,N))
v_sources = numpy.zeros((N,N))
psi_sources = numpy.zeros((N,N))

for i in index:

    us, vs = get_velocity_sources(strength_sources[i], xs_airfoil[i], ys_airfoil[i], X, Y)

    u_sources += us
    v_sources += vs

    psi = get_streamfunction_sources(strength_sources[i], xs_airfoil[i], ys_airfoil[i], X, Y)

    psi_sources += psi
```

### Determine velocities by superposition

```python
# superpose
u = u_sources + u_inf*numpy.ones((N,N))  #creates a 51 by 51 matrix with u_inf values
v = v_sources
```

```python
# plots the streamlines
%matplotlib inline

size = 20
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(xs_airfoil, ys_airfoil, color='#CD2305', s=50, marker='o');
```

Not that the plot above shows the sources in red, which follow the outline of the airfoil. 

## Compute the Pressure Coefficient

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2
```

```python
# plots the pressure coefficient field
size = 20
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-0.6, .6, 20), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])

pyplot.scatter(xs_airfoil, ys_airfoil, color='k', s=50, marker='o');

```

```python
# This line of code shows the array indices for maximum Cp:
numpy.unravel_index(numpy.argmax(cp), cp.shape)
```

```python
# This line of code shows the maximum value of Cp (cp[25,16])
numpy.max(cp)
```

Also, note that as N is increased, the max CP number approaches the actual max theoretical value of CP =1. This is a good example of mesh refinement to obtain the expected results. Try with N=1000 and you will see that CP is about CP = 0.9
