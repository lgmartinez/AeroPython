# Lift on a Cylinder

First determine flow around a cylinder (using the uniform flow past a doublet from Lesson 4)

### Import Libraries

```python
import math
import numpy
from matplotlib import pyplot
```

### Create Grid

```python
N = 50

x_start, x_end = -2.0, 2.0
y_start, y_end = -1.0, 1.0

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x, y)
```

### Strength and location of doublet

```python
strength_doublet = 1.0

x_doublet, y_doublet = 0.0, 0.0
```

### Freestream Velocity

```python
u_inf = 1.0
```

### Define functions to calculate velocities and streamfunction

```python
def get_velocity_doublet(strength, xd, yd, X, Y):

    u = - strength/(2*math.pi)*((X-xd)**2-(Y-yd)**2)/((X-xd)**2+(Y-yd)**2)**2
    v = - strength/(2*math.pi)*2*(X-xd)*(Y-yd)/((X-xd)**2+(Y-yd)**2)**2
    
    return u, v
    
def get_streamfunction_doublet(strength, xd, yd, X, Y):

     psi = - strength/(2*math.pi)*(Y-yd)/((X-xd)**2+(Y-yd)**2)
        
     return psi
```

### Compute velocities and streamfunctions

```python
# Doublet velocities

u_doublet, v_doublet = get_velocity_doublet(strength_doublet, x_doublet, y_doublet, X, Y)


# Doublet Streamfunction

psi_doublet = get_streamfunction_doublet(strength_doublet, x_doublet, y_doublet, X, Y)

# Freestream velocities

u_freestream = u_inf*numpy.ones((N,N), dtype=float)
v_freestream = numpy.zeros((N,N), dtype=float)

# Freestream streamfunction

psi_freestream = u_inf*Y

# Superposed velocities and streamfunction

u = u_freestream + u_doublet
v = v_freestream + v_doublet

psi = psi_doublet + psi_freestream
```

### Stagnation Points

```python
#These will be at locations where u = 0, where y = 0, then solve for x locations:

x_stagn1, y_stagn1 = +math.sqrt(strength_doublet/(2*math.pi*u_inf)), 0
x_stagn2, y_stagn2 = -math.sqrt(strength_doublet/(2*math.pi*u_inf)), 0
```

### Radius of Cylinder

```python
# The radius of the cylinder is the absolute value distance from a stagnation point to the location of the doublet (the origin):

R = math.sqrt(strength_doublet/(2*math.pi*u_inf))
```

### Plot!

```python
%matplotlib inline

# plots the streamlines
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_doublet, y_doublet, color='K', s=200, marker='*')

# Add Cylinder surface area (circle)
circle = pyplot.Circle((0, 0), radius=R, color='r', alpha=0.5)
pyplot.gca().add_patch(circle)

# Add stagnation points to graph
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn1, y_stagn2], color='g', s=80, marker='o');
```

Now, to examine lift on the cylinder we add a vortex in the center of the cylinder

### Strength and Location of Vortex

```python
strength_vortex = 4.0    # strength of vortex

x_vortex, y_vortex = 0.0, 0.0  #location of vortex
```

### Define functions for velocities and streamfunction

```python
def get_velocity_vortex(strength, xv, yv, X, Y):

    u = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u, v
    
def get_streamfunction_vortex(strength, xv, yv, X, Y):
    
    psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return psi
    
```

### Compute Vortex Velocities and Streamfunction

```python
u_vortex, v_vortex = get_velocity_vortex(strength_vortex, x_vortex, y_vortex, X, Y)

psi_vortex = get_streamfunction_vortex(strength_vortex, x_vortex, y_vortex, X, Y)
```

### Superposed Velocities and streamfunction

```python
u = u_freestream + u_doublet + u_vortex
v = v_freestream + v_doublet + v_vortex
psi = psi_freestream + psi_doublet + psi_vortex
```

### Stagnation Points

In order to find the stagnation points we set the superposed components of velocity equal to zero

From v = 0, we find that y_stagnation = -(Gamma*R^2)/(2*Kappa)

Since we kow that R = sqrt (Kappa/(2*pi*u_inf)), (we know R from the radius creeated by doublet) we can rewrite y as

y_stagnation = Gamma/(4*pi*u_inf)

then we find x_stagnation points by

x_stagnation^2 + y_stagnation = R^2, where

x_stagnadtion = +/- sqrt(R^2 - (Gamma/(4*pi*u_inf))^2)

```python
x_stagn1 = + numpy.sqrt(R**2 - (strength_vortex/(4*math.pi*u_inf))**2)
x_stagn2 = - numpy.sqrt(R**2 - (strength_vortex/(4*math.pi*u_inf))**2)

y_stagn = - strength_vortex/(4*math.pi*u_inf)  # note that y_stagn1=y_stagn2
```

### Plot!

```python
size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))

pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)

pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)

pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1.5, arrowstyle='->')

circle = pyplot.Circle((0, 0), radius=R, color='r', alpha=0.5)
pyplot.gca().add_patch(circle)

pyplot.scatter(x_vortex, y_vortex, color='r', s=80, marker='o')
pyplot.scatter([x_stagn1, x_stagn2], [y_stagn, y_stagn], color='g', s=80, marker='o');
```

### Pressure Coefficient

### Calculate CP: with and without vortex

```python
theta = numpy.linspace(0,4*math.pi, 100)

u_no_vortex = - 2*u_inf*numpy.sin(theta)
u_with_vortex = - 2*u_inf*numpy.sin(theta) - (strength_vortex/(2*math.pi*R))

# CP due to uniform stream + doublet

CP_no_vortex = 1 - (u_no_vortex/u_inf)**2

# CP due to uniform stream + doublet + vortex

CP_with_vortex = 1 - (u_with_vortex/u_inf)**2
```

### Plot CP

```python
size = 10

pyplot.figure(figsize=(size, size))
pyplot.grid(True)

pyplot.xlabel('$theta$', fontsize=18)
pyplot.ylabel('$C_p$', fontsize=18)

pyplot.xlim(theta.min(), theta.max())

pyplot.plot(theta, CP_with_vortex, color='m', linewidth=2, linestyle='-')
pyplot.plot(theta, CP_no_vortex, color='g', linewidth=2, linestyle='-')

pyplot.legend(['with vortex', 'without vortex'], loc='best', prop={'size':16});
```
