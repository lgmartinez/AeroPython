### Part 1: Circle at origin with radius R =1 

### Xi and Eta Coordinates

```python
N=21

xi_start, xi_end = -1.0, 1.0

xi = numpy.linspace(xi_start, xi_end, N)

eta = numpy.sqrt(xi_end-(xi)**2)
eta2 = - numpy.sqrt(xi_end-(xi)**2)
```

```python
%matplotlib inline

plot_limit_xi_start = xi_start - 0.5
plot_limit_xi_end = xi_end + 0.5

pyplot.plot(xi, eta, 'o', color = 'k')
pyplot.plot(xi, eta2, 'o', color = 'k')
pyplot.axis('equal')

pyplot.xlim(plot_limit_xi_start, plot_limit_xi_end)
pyplot.ylim(plot_limit_xi_start, plot_limit_xi_end);
```

```python
ETA=numpy.array([eta,eta2])

XI=numpy.array([xi,xi])
```

```python
#A Python complex number z is stored internally using rectangular or Cartesian coordinates. 
#It is completely determined by its real part z.real and its imaginary part z.imag. In other words:

zeta = XI + ETA*1j
```

```python
z = zeta + (1/zeta)
```

```python
x = numpy.real(z)
y = numpy.imag(z)
```

```python
%matplotlib inline

plot_limit_x_start = numpy.min(x)-1
plot_limit_x_end = numpy.max(x)+1

plot_limit_y_start = numpy.min(y)-1
plot_limit_y_end = numpy.max(y)+1

pyplot.plot(x, y, 'o', color = 'k')
pyplot.plot(x, y, 'o', color = 'k')
#pyplot.axis('equal')
pyplot.ylim(plot_limit_y_start, plot_limit_y_end)
pyplot.xlim(plot_limit_x_start, plot_limit_x_end);
```

## Flow Past Cylinder: Doublet + Freestream

get the value of streamfunction on a grid of points for flow around the cylinder of (3); now map those points to the z plane ... plot the locations of the mapped points

### Create Grid 

```python
N = 51

x_start, x_end = -3.0, 3.0
y_start, y_end = -2.0, 2.0

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X,Y = numpy.meshgrid(x, y)
```

```python

pyplot.scatter(X, Y, s=10, color='k', marker='o', linewidth=0);  
```

### Doublet Location and strength

```python
x_doublet, y_doublet = -0.5, 0.0

strength_doublet = 2*math.pi*(R**2)
```

### Doublet Velocities and Streamfunction

```python
def get_velocity_doublet(strength, x_d, y_d, X, Y):
    "Returns the velocity field generated by doublet"
    
    u = - strength/(2*math.pi)*((X-x_d)**2-(Y-y_d)**2)/((X-x_d)**2+(Y-y_d)**2)**2
    v = - strength/(2*math.pi)*2*(X-x_d)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)**2
    
    return u, v

def get_streamfunction_doublet(strength, x_d, y_d, X, Y):
    "Returns the stream function generated by doublet"
    
    psi = - strength/(2*math.pi)*(Y-y_d)/((X-x_d)**2+(Y-y_d)**2)
    
    return psi
```

```python
u_doublet, v_doublet = get_velocity_doublet(strength_doublet, x_doublet, y_doublet, X, Y)

psi_doublet = get_streamfunction_doublet(strength_doublet, x_doublet, y_doublet, X, Y)
```

### Freestream Velocities and Streamfunction

```python
u_inf = 1.0  # free stream speed

u_freestream = u_inf*numpy.ones((N,N), dtype=float)
v_freestream = numpy.zeros((N,N),dtype=float)

psi_freestream = u_inf * Y
```

### Superposed Velocities and Streamfunction

```python
u = u_freestream + u_doublet
v = v_freestream + v_doublet

psi = psi_freestream + psi_doublet
```

### Graph

```python
%matplotlib inline


size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)

pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.contour(X, Y, psi, levels=[0.], colors='r', linewidths=2, linestyles='solid')
pyplot.scatter(x_doublet, y_doublet, color='K', s=160, marker='*');
```

### Transformation to z plane

```python
ZETA = X + Y*1j
zeta = x + y*1j
```

```python
Z = ZETA + (1/ZETA)
z = zeta + (1/zeta)
```

```python
X_zplane = numpy.real(Z)
x_zplane = numpy.real(z)

Y_zplane = numpy.imag(Z)
y_zplane = numpy.imag(z)
```

```python
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.scatter(X_zplane, Y_zplane, s=10, color='k', marker='o', linewidth=0);
```

This is what the transformation of the XI and ETA gridpoints looks like! They are all concentrated in one rectangular area.
