## Part 3 Grid points on the z-plane using polar coordinates

```python
xc, yc = -0.15, 0.0
N_th = 145
th = numpy.linspace(0, 2*numpy.pi, N_th)
```

```python
N_r = 100
r_start, r_end = 1.15, 5.0
r = numpy.linspace(r_start, r_end, N_r)
```

```python
z = numpy.empty((N_r, N_th), dtype=numpy.complex128)

for i in range(N_r):
    for j in range(N_th):
        z[i,j] = r[i] * numpy.exp(th[j] * 1j) + xc
```

```python
x = numpy.real(z)
y = numpy.imag(z)
```

```python
xi = z + ((c**2)/z)
zeta = numpy.real(xi)
eta = numpy.imag(xi)
```

```python
pyplot.figure(figsize=(10,10))

pyplot.subplot(2,2,1)
pyplot.axis('equal')
pyplot.scatter(x,y,color='k', s=0.4, marker='o')
pyplot.title('Z-Plane');

pyplot.subplot(2,2,2)
pyplot.axis('equal')
pyplot.scatter(zeta,eta,color='k', s=0.4, marker='o')
pyplot.title('Xi-Plane');

```

## Part 4

### Now, you will evaluate potential flow over the cylinder on the z-plane. As mentioned above, ψ(ξ)=ψ(ξ(z)). This means that, after evaluating the stream function at a certain point on the z-plane, the corresponding point on the ξ-plane has the same value of the stream function. We can plot the streamlines on both planes using the function pyplot.contour(), because the stream function is a scalar function.  Use 1 as the free stream velocity, i.e., U∞=1. You have to calculate the strength of the doublet first in order to have a cylinder with radius R=1.15.

```python
u_inf = 1.0
R = 1.15
```

```python
#strength of doubtlet:
k = numpy.pi*2*u_inf*(R**2)
```

```python
print 'The strength of the doublet is ', numpy.round(k,2)
```

```python
# Doublet Streamfunction:

psi_doublet = numpy.empty((N_r, N_th), dtype=numpy.complex128)

for i in range(N_r):
    for j in range(N_th):
        psi_doublet[i,j] = -1*(k*numpy.sin(th[j]))/(2*numpy.pi*r[i]) 
```

```python
#Freestream Streamfunction:

psi_freestream = numpy.empty((N_r, N_th), dtype=numpy.complex128)

for i in range(N_r):
    for j in range(N_th):
        psi_freestream[i,j] = u_inf*r[i]*numpy.sin(th[j])
```

```python
# superposition of streamfunction

psi = psi_doublet + psi_freestream
```

```python
contour = numpy.linspace(numpy.min(psi),numpy.max(psi), (N_r/2)+1)
print len(contour)
```

```python
pyplot.figure(figsize=(10,10))

pyplot.subplot(2,2,1)
pyplot.axis('equal')
pyplot.contour(x, y, psi, levels=contour, colors='k', linestyles='solid')
pyplot.plot(x[0],y[0])
pyplot.title('Streamlines in Z-Plane');

pyplot.subplot(2,2,2)
pyplot.axis('equal')
pyplot.contour(zeta, eta, psi, levels=contour, colors='k', linestyles='solid')
pyplot.plot(zeta[0],eta[0])
pyplot.title('Streamlines in Xi-Plane');
```

### Part 4 - Continued - Velocity vectors and pressure coefficients

### Velocities in z-plane:

```python
# doublet velocitites:
ud_z = - k/(2*numpy.pi)*((x-xc)**2-(y-yc)**2)/((x-xc)**2+(y-yc)**2)**2
vd_z = - k/(2*numpy.pi)*2*(x-xc)*(y-yc)/((x-xc)**2+(y-yc)**2)**2
```

```python
#Freestream velocities
u_z_freestream = u_inf*numpy.ones((N_r,N_th), dtype=float)
v_z_freestream = numpy.zeros((N_r,N_th), dtype=float)
```

```python
#Superposition:
uz = ud_z + u_z_freestream
vz = vd_z + v_z_freestream
```

### velocities in the xi plane

```python
dF_dz = uz -(1j*vz)
dxi_dz = 1 - (c/z)**2

w_xi = dF_dz/dxi_dz
```

```python
u_xi = numpy.real(w_xi)
v_xi = -1*numpy.imag(w_xi)
```

```python
step = 5
pyplot.figure(figsize=(15,15))

pyplot.subplot(2,2,1)
pyplot.axis('equal')
pyplot.quiver(x[::step],y[::step],uz[::step],vz[::step],angles='uv',scale=33)
pyplot.xlim(-3.0,3.0)
pyplot.ylim(-3.0,3.0)
pyplot.title('Velocity Vector Plot in Z-Plane');

pyplot.subplot(2,2,2)
pyplot.axis('equal')
pyplot.quiver(zeta[::step],eta[::step],u_xi[::step],v_xi[::step],angles='uv',scale=33)
pyplot.xlim(-3.0,3.0)
pyplot.ylim(-3.0,3.0)
pyplot.title('Velocity Vector Plot in Xi-Plane');
```

### Pressure coefficient in z-plane and xi-plane

```python
# computes the pressure coefficient field in z-plane
cp_z = 1.0 - (uz**2+vz**2)/u_inf**2

# computes the pressure coefficient field in xi-plane
cp_xi = 1.0 - (u_xi**2+v_xi**2)/u_inf**2
```

```python
# plots the pressure coefficient field

pyplot.figure(figsize=(15,15))

pyplot.subplot(2,2,1)
pyplot.axis('equal')
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(-6.0, 6.0)
pyplot.ylim(-6.0, 6.0)
contf = pyplot.contourf(x, y, cp_z, levels=numpy.linspace(-1.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-0.88,-0.66,-0.44,-0.22,0.00,0.22,0.44,0.66,0.88])
pyplot.contour(x,y,psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid');
pyplot.title('CP in Z-Plane')

pyplot.subplot(2,2,2)
pyplot.xlabel('zeta', fontsize=16)
pyplot.ylabel('eta', fontsize=16)
pyplot.xlim(-6.0, 6.0)
pyplot.ylim(-6.0, 6.0)
contf = pyplot.contourf(zeta, eta, cp_xi, levels=numpy.linspace(-1.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-0.88,-0.66,-0.44,-0.22,0.00,0.22,0.44,0.66,0.88])
#pyplot.scatter(x_doublet, y_doublet, color='#CD2305', s=80, marker='o')
pyplot.contour(zeta,eta,psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid');
pyplot.title('CP in Xi-Plane');
```

## Part 6 - Flow over a symmetric Joukowski airfoil with non-zero angle of attack, with circulation

stagnation point will occur at trailing edge of airfoil; based on coordinate system this occurs at [zeta(max), eta(min)]

```python
x_stagn1 = R+xc
G = -4*numpy.pi*u_inf*numpy.sin(AoA2_rad)*(x_stagn1-xc)
```

```python
print 'The strength of the vortex is', numpy.round(G,2)
```

```python
# vortex streamfunction in z prime plane:
psi_prime_vortex = (G/(4*numpy.pi))*numpy.log((x_prime-xc_prime)**2+(y_prime-yc_prime)**2)
```

```python
# Superposition of streamfunction in z prime plane:
psi_prime_with_vortex = psi_prime + psi_prime_vortex 
```

```python
# vortex velocities in z prime plane:
uz_prime_vortex = + G/(2*numpy.pi)*((y_prime-yc_prime)/((x_prime-xc_prime)**2+(y_prime-yc_prime)**2))
vz_prime_vortex = - G/(2*numpy.pi)*((x_prime-xc_prime)/((x_prime-xc_prime)**2+(y_prime-yc_prime)**2))
```

```python
# superposition of velocities in z prime plane:
uz_prime_withvortex = uz_prime + uz_prime_vortex
vz_prime_withvortex = vz_prime + vz_prime_vortex
```

```python
# Transformation of velocities to Z plane:
uz_withvortex = uz_prime_withvortex*numpy.exp(-1j*AoA_rad)
vz_withvortex = vz_prime_withvortex*numpy.exp(-1j*AoA_rad)
```

```python
# Transformation to Xi plane:
dF_dz_wv = uz_withvortex -(1j*vz_withvortex)
dxi_dz_wv = 1 - (c/z)**2

w_xi_wv = dF_dz_wv/dxi_dz_wv

u_xi_withvortex = numpy.real(w_xi_wv)
v_xi_withvortex = -1*numpy.imag(w_xi_wv)
```

```python
contour1 = numpy.linspace(numpy.min(psi_prime_with_vortex),numpy.max(psi_prime_with_vortex), (N_r/2)+1)

pyplot.figure(figsize=(12,12))

pyplot.subplot(2,2,1)
pyplot.axis('equal')
pyplot.contour(x, y, psi_prime_with_vortex, levels=contour1, colors='k', linestyles='solid')
pyplot.plot(x[0],y[0])
pyplot.title('Z-Plane')

pyplot.subplot(2,2,2)
pyplot.axis('equal')
pyplot.contour(zeta, eta, psi_prime_with_vortex, levels=contour1, colors='k', linestyles='solid')
pyplot.plot(zeta[0],eta[0])
pyplot.title('Xi-Plane');
```

### Stagnation Points

```python
xprime_stagn1, xprime_stagn2  = + numpy.sqrt(R**2 - (G/(4*math.pi*u_inf))**2), - numpy.sqrt(R**2 - (G/(4*math.pi*u_inf))**2)
yprime_stagn1, yprime_stagn2 = - G/(4*math.pi*u_inf), - G/(4*math.pi*u_inf)  
```

```python
#AoA1, AoA2 = 20.0, 200 
#AoA1_rad, AoA2_rad = ((AoA1 * numpy.pi)/180), ((AoA2 * numpy.pi)/180)
#factor1 = numpy.cos(AoA1_rad) + numpy.tan(AoA1_rad)*numpy.sin(AoA1_rad)
#factor2 = numpy.cos(AoA2_rad) + numpy.tan(AoA2_rad)*numpy.sin(AoA2_rad)

x_stagn1, x_stagn2 = ((xprime_stagn1/factor1) + xc), ((xprime_stagn1/factor2) + xc)
y_stagn1, y_stagn2 = (x_stagn1-xc)*numpy.tan(AoA1_rad), (x_stagn2-xc)*numpy.tan(AoA2_rad)
```

```python
# Transform these to the xi-plane:
z_stagn1 = x_stagn1 + 1j*y_stagn1
xi_stagn1 = z_stagn1 + ((c**2)/z_stagn1)
zeta_stagn1 = numpy.real(xi_stagn1)
eta_stagn1 = numpy.imag(xi_stagn1)

z_stagn2 = x_stagn2 + 1j*y_stagn2
xi_stagn2 = z_stagn2 + ((c**2)/z_stagn2)
zeta_stagn2 = numpy.real(xi_stagn2)
eta_stagn2 = numpy.imag(xi_stagn2)
```
