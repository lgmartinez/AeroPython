## Import Libraries

```python
import numpy
import math
from matplotlib import pyplot
```

## Define Grid

```python
N=100

x_start, x_end = -5.0, 5.0
y_start, y_end = -2.0, 2.0

x = numpy.linspace(x_start, x_end, N)
y = numpy.linspace(y_start, y_end, N)

X, Y = numpy.meshgrid(x,y)
```

vortices will all be of equal strength (try to create code with vortices of different strengths after completing this code)

## Location of vortices

This code will use a total of 10 vortices, equally spaced on the x-axis from the origin to +/- x

```python
# This shows the x-location of vortices

xv_list = numpy.arange(-5,6)

print xv_list
```

```python
# This shows y-location of vortices

yv_list = numpy.zeros(11)

print yv_list
```

## Strength of Vortices

```python
G = 5.0

gamma = numpy.repeat(G, numpy.size(xv_list))

print gamma
```

```python
index = numpy.arange(0,len(xv_list))



print index
```

## Define functions to calculate velocities and stream function

```python
def get_velocity_vortices(strength, xv, yv, X, Y):

    u_vortex = + strength/(2*math.pi)*(Y-yv)/((X-xv)**2+(Y-yv)**2)
    v_vortex = - strength/(2*math.pi)*(X-xv)/((X-xv)**2+(Y-yv)**2)
    
    return u_vortex, v_vortex

def get_streamfunction(strength, xv, yv, X, Y):

    psi = psi = strength/(4*math.pi)*numpy.log((X-xv)**2+(Y-yv)**2)
    
    return psi
```

## Calculate velocities and streamfunction

```python

u_vortices = numpy.zeros((N,N))
v_vortices = numpy.zeros((N,N))

psi_vortices = numpy.zeros((N,N))

for i in index:
    
    u_v, v_v = get_velocity_vortices(gamma[i], xv_list[i], yv_list[i], X, Y)
    
    u_vortices += u_v
    v_vortices += v_v
    
    psi = get_streamfunction(gamma[i],xv_list[i], yv_list[i], X, Y)
    
    psi_vortices +=psi
```

## Plot

```python
# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u_vortices, v_vortices, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(xv_list, yv_list, color='#CD2305', s=40, marker='o');
```
