# Source & Sink in a Freestream

## Import Librarues and Create Grid

```python
import numpy                    #Imports array library
from matplotlib import pyplot   #Imports plotting functions
import math                     #Imports mathematical functions
```

```python
N = 200                               # Number of points in each direction
x_start, x_end = -4.0, 4.0            # x-direction boundaries
y_start, y_end = -2.0, 2.0            # y-direction boundaries
x = numpy.linspace(x_start, x_end, N)    # 1D-array for x
y = numpy.linspace(y_start, y_end, N)    # 1D-array for y
X, Y = numpy.meshgrid(x, y)              # generates a mesh grid

#Note that "numpy.shape(X)" will display the array size (also works for Y)

numpy.shape(X)
```

## Source in a Uniform Flow

```python
u_inf = 1.0        # freestream speed

# computes the freestream velocity field
u_freestream = u_inf * numpy.ones((N, N), dtype=float)
v_freestream = numpy.zeros((N, N), dtype=float)

# More on data type object, dtype: http://docs.scipy.org/doc/numpy/reference/arrays.dtypes.html

# computes the stream-function
psi_freestream = u_inf * Y
```

Let's define a function called get_velocity() to compute the velocity components (u,v) on a mesh (X,Y), given the strength of the singularity and its location (xs,ys). We define it with the def command, followed by our chosen name for the function, and any parameters inside the parenthesis. The line ends with a colon:

```python
def get_velocity(strength, xs, ys, X, Y):
    """Returns the velocity field generated by a source/sink.
    
    Arguments
    ---------
    strength -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid.
    """
    u = strength/(2*math.pi)*(X-xs)/((X-xs)**2+(Y-ys)**2)
    v = strength/(2*math.pi)*(Y-ys)/((X-xs)**2+(Y-ys)**2)
    
    return u, v
```

```python
help(get_velocity)
```

```python
def get_streamfunction(strength, xs, ys, X, Y):
    """ Returns the stream-function generated by a source/sink.
    
    Arguments
    ---------
    strenght -- strength of the source/sink.
    xs, ys -- coordinates of the source/sink.
    X, Y -- mesh grid
    """
    psi = strength/(2*math.pi)*numpy.arctan2((Y-ys), (X-xs))
    
    return psi
```

### Source Velocities and Streamfunction

```python
strength_source = 5.0
x_source, y_source = -1.0, 0.0

# to compute velocity field:
u_source, v_source = get_velocity(strength_source, x_source, y_source, X, Y)

# to compute the streamfunction
psi_source = get_streamfunction(strength_source, x_source, y_source, X, Y)
```

### Plot Streamlines

```python
# superposition of the source on the freestream

u = u_freestream + u_source
v = v_freestream + v_source
psi = psi_freestream + psi_source

# plots the streamlines
%matplotlib inline

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.grid(True)
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter(x_source, y_source, color='r', s=80, marker='o')

# to calculate stagnation point:
x_stag = x_source - strength_source/(2*math.pi*u_inf)
y_stag = y_source

# add stagnation points to graph:
pyplot.scatter(x_stag, y_stag, color='g', s=80, marker='o')

# add dividing line across x-axis:
pyplot.contour(X, Y, psi, levels=[-strength_source/2, +strength_source/2], colors='r', linewidths=2, linestyles='solid');
```

## Source-Sink Pair in Unifrom Flow

```python
strength_sink = -5.0
x_sink, y_sink = 1.0, 0.0

# to compute the velocity field on the grid:
u_sink, v_sink = get_velocity(strength_sink, x_sink, y_sink, X, Y)

# to compute the streanfunction on the grid:
psi_sink = get_streamfunction(strength_sink, x_sink, y_sink, X, Y)
```

```python
# By superposition:
u = u_freestream + u_source + u_sink
v = v_freestream + v_source + v_sink
psi = psi_freestream + psi_source + psi_sink
```

### Plot Streamlines

```python
# Plot the streamlines:

size = 10
pyplot.figure(figsize=(size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
pyplot.streamplot(X, Y, u, v, density=2, linewidth=1, arrowsize=1, arrowstyle='->')
pyplot.scatter([x_source, x_sink], [y_source, y_sink], color='r', s=80, marker='o')
pyplot.contour(X, Y, psi, levels=[0.], colors='r', linewidths=2, linestyles='solid');
```

## Bernoulli's Equation and Pressure Coefficient

```python
# computes the pressure coefficient field
cp = 1.0 - (u**2+v**2)/u_inf**2

# plots the pressure coefficient field
size = 10
pyplot.figure(figsize=(1.1*size, (y_end-y_start)/(x_end-x_start)*size))
pyplot.xlabel('x', fontsize=16)
pyplot.ylabel('y', fontsize=16)
pyplot.xlim(x_start, x_end)
pyplot.ylim(y_start, y_end)
contf = pyplot.contourf(X, Y, cp, levels=numpy.linspace(-2.0, 1.0, 100), extend='both')
cbar = pyplot.colorbar(contf)
cbar.set_label('$C_p$', fontsize=16)
cbar.set_ticks([-2.0, -1.0, 0.0, 1.0])
pyplot.scatter([x_source, x_sink], [y_source, y_sink], color='#CD2305', s=80, marker='o')
pyplot.contour(X, Y, psi, levels=[0.], colors='#CD2305', linewidths=2, linestyles='solid');
```

The areas of highes pressure are near the stagnation points, as velocity is the lowest in this area. According to Bernoulli's equation, we see that at the location when v --> 0, pressure will be highest at that location. This is what we see in the contour plot above.
